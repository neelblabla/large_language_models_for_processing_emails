```python
import pandas as pd
import re
```

```python
df=pd.read_csv('enron_mails_subsample.csv')
df
```

```python
labels=[]
with open('1.txt', 'r') as file:
    for line in file:
        li=line.strip()
        print(li)
        li=re.findall('(^.*?):',li)
        labels.append(li)
label=[i for sl in labels for i in sl]
label=label[:-1]
# label.append('Body')
label
```

```python
for a in label:
    if not df['RawMailContent'].str.contains(a).any():
        continue
    #if df['RawMailContent'].str.extract(a+r':\s*(.*)').iloc[0,0] in label:
     #   continue
    df[a]=df['RawMailContent'].str.extract(a+r':\s(.*)')
    
    df['RawMailContent']=df['RawMailContent'].str.replace(a+r':\s(.*)','')
    df['RawMailContent']=df['RawMailContent'].str.strip('\n')
df
```
