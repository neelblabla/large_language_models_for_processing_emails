```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_path = "/Users/nilsmart96/Code/fs/spm/fs-spm-project/llama_fine_hf" # Specify folder where the model files from huggingface are located
model = AutoModelForCausalLM.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
prompt = "what happened once upon a time?"

input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)
output_ids = model.generate(input_ids, max_length=25, num_return_sequences=1, no_repeat_ngram_size=2)
generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(generated_text)
```
