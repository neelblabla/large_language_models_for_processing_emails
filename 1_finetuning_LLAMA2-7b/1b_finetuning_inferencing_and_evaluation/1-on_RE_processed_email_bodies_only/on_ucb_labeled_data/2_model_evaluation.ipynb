### Assessing and evaluating model outputs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import model_test_output-evaluation_df.csv in a dataframe
model_output_df = pd.read_csv("model_test_output-evaluation_df.csv")
model_output_df.head()
```

```python
# pie-chart of model response 'match' with human labels 
total = model_output_df['match'].value_counts().values.sum()
def fmt(x):
    return '{:.1f}%\n{:.0f}'.format(x, total*x/100)
plt.pie(model_output_df['match'].value_counts().values, labels=model_output_df['match'].value_counts().index, autopct=fmt)
```

```python
# Plotting training loss data

training_loss_data = {
    "Step": [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325],
    "Training Loss": [
        2.305800, 0.401000, 1.525900, 0.369800, 1.451300, 0.482000,
        1.369700, 0.311900, 1.327900, 0.300900, 1.348400, 0.279400, 1.024500
    ],
}

training_loss_df = pd.DataFrame(training_loss_data)

plt.figure(figsize=(10, 6))
plt.plot(training_loss_df["Step"], training_loss_df["Training Loss"], marker='o', linestyle='-')
plt.title("Training Loss during 'peft fine-tuning llama2-7b'")
plt.xlabel("Step")
plt.ylabel("Training Loss")
plt.grid(True)

plt.tight_layout()
plt.show()
```
