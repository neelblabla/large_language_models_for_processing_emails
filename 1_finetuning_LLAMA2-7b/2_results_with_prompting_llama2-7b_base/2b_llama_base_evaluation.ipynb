```python
import pandas as pd
```

```python
df = pd.read_csv("llama_test.csv")
```

```python
df.head()
```

```python
df = df[["cat_1_level_2", "llama output"]]

df.head()
```

```python
# Replace "1." with nothing in the 'llama output' column (and other special cases)
df['llama output'
  ] = df['llama output'
        ].str.replace('Therefore, categories 1.1, 1.2, and 1.4 are not applicable.', '', regex=False)
df['llama output'
  ] = df['llama output'
        ].str.replace('The categories 1.2, 1.3, 1.4, and 1.5 do not seem', '', regex=False)
df['llama output'] = df['llama output'].str.replace('1.2-1.5 do not fit', '', regex=False)
df['llama output'] = df['llama output'].str.replace('SoCal Edison in RP00-241,', '', regex=False)

df['llama output'] = df['llama output'].str.replace('1\.', '', regex=True)
df['llama output'] = df['llama output'].str.replace('Section 3', '', regex=False)

# Drop rows with NaN values in the "llama_output" column
df = df.dropna(subset=['llama output'])

df
```

```python
# Convert the 'cat_1_level_2' column to integers, then to strings
df['cat_1_level_2'] = df['cat_1_level_2'].astype(int)
df = df.astype(str)
# Check if the number from the first column is found in the second column
df['correct label'] = df.apply(lambda row: row['cat_1_level_2'] in row['llama output'], axis=1)

df.head(20)
```

```python
df.to_csv("llama_acc_test.csv")
```
