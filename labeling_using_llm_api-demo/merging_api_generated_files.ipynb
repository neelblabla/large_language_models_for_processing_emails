```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('GPT_api_label.csv')
df.drop(['Unnamed: 0','Body'],axis=1,inplace=True)
df
```

```python
df1=pd.read_csv('enron_mails_subsample.csv')
df1.drop('Unnamed: 0',axis=1,inplace=True)
df1
```

```python
df=pd.merge(df, df1, on='Message-ID', how='left')
```

```python
df
```

```python
df=df[~((df['content'].isna()) & (df['subject'].isna()))]
df
```

```python
df['result']=df['result'].str[0]
df['result'].unique()
```

```python
df=df[~(df['result']=='P')]
df=df[~(df['result']=='C')]
df=df[~(df['result']=='L')]

df['result'].unique()
```

```python
df.to_csv('GPT3.5_output.csv')
```
